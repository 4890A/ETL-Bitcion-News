```python
import pandas as pd
import numpy as np
import time
```

```python
historical_df = pd.read_csv('BTC_Price_2013_2019.csv')
historical_df.head()
```

```python
historical_df['date'] = pd.to_datetime(historical_df['timestamp'])
historical_df['date'] = historical_df['date'].dt.date
historical_df = historical_df[["date", "open", "close", "Δprice", "volume"]]
historical_df.head()
```

```python
sentiment_df = pd.read_csv('articles_with_sentiment.csv')
sentiment_df = sentiment_df[["date", "headline", "abstract", "sentiment", "url"]]
sentiment_df.head()
```

```python
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])
sentiment_df['date'] = sentiment_df['date'].dt.date
sentiment_df.head()
```

```python
merged_df = sentiment_df.merge(how='left', right=historical_df, left_on='date', right_on='date')
merged_df
```

```python
trend_df = pd.read_csv('Google_Trends.csv')
trend_df = trend_df[:1961]
trend_df.rename(columns={"Unnamed:0": "date", "Category:All categories": })
```

```python
# Save dataframe as CSV
merged_df.to_csv('BTC_Price_Sentiment.csv')

# Save dataframe as JSON
merged_df.to_json('BTC_Price_Sentiment.json')
```
